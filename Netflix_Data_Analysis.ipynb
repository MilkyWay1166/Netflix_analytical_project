## IV. PREPARING DATA

```python
data['main_genre'] = data['listed_in'].apply(lambda x: x.split(',')[0].strip())
data['main_genre']
```

```python
data['year_added'] = data['date_added'].astype('str').apply(lambda x: x.split(' ')[-1])
```

```python
data['country'] = data['country'].str.strip().str.split(', ')
data['cast'] = data['cast'].str.strip().str.split(', ')
data['director'] = data['director'].str.strip().str.split(', ')
data['listed_in'] = data['listed_in'].str.strip().str.split(', ')
```

There are so many rating categories in the column [rating] and not easy to get their meaning. For easier understanding, I translate them into target audiences which should be a clearer angle for analyzing.

```python
ratings_ages = {
    'TV-PG': 'Teenagers',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Kids',
    'TV-Y7': 'Kids',
    'TV-14': 'Teenagers',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Unrated',
    'PG-13': 'Teenagers',
    'TV-G': 'General Audience',
    'PG': 'Teenagers',
    'G': 'General Audience',
    'UR': 'Unrated',
    'NC-17': 'Adults'
}
target_audience = list(ratings_ages.values())
data['target_audience'] = data['rating'].replace(ratings_ages).map(lambda x: x if x in target_audience else 'Unrated')
data['target_audience']
```

## 2. Top 10 Content Production Countries on Netflix

USA is significantly the top 1 contender. 

Most countries in top 10 tend to produce more Movies.

Japan and South Korea are relatively different from others that more than 50% of their show are TV Shows

```python

#Explode Country column
xploded_data = data[['country','type']].explode('country')

# Find top 10 countries
top10_countries = xploded_data[xploded_data['country'].notnull()]['country']\
                    .value_counts()[:10]\
                    .index#reset_index(name = 'counts')

top10_countries_groupby = xploded_data[xploded_data['country'].isin(top10_countries)] \
                        .groupby(['country','type']).size()\
                        .reset_index(name = 'No. shows')

top10_countries_groupby['Total shows'] = top10_countries_groupby[['country','No. shows']].groupby('country').transform('sum')
top10_countries_groupby['Contribution'] = top10_countries_groupby['No. shows']/top10_countries_groupby['Total shows']*100

fig, ax = plt.subplots(1,2,figsize = (19,7))

# stacked barh chart
ax[0].barh(width= 'No. shows', y = 'country', 
        data = top10_countries_groupby[top10_countries_groupby['type']=='Movie'].sort_values('Total shows', ascending=True), 
        left=top10_countries_groupby[top10_countries_groupby['type']=='TV Show'].sort_values('Total shows', ascending=True)['No. shows'], 
        color = '#221f1f',
        label = 'Movie')

ax[0].barh(width = 'No. shows', y = 'country', 
        data = top10_countries_groupby[top10_countries_groupby['type']=='TV Show'].sort_values('Total shows', ascending=True),
        color = '#b20710',
        label = 'TV shows')
#ax[0].set_title('No. Movies & TV Shows across Top 10 Countries')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].invert_xaxis()

for i, v in enumerate(top10_countries_groupby[top10_countries_groupby['type']=='Movie'].sort_values('Total shows', ascending=True)['Total shows']):
    ax[0].text(v + 50, i, str(v), va='center', ha='right', fontsize=11, c = 'black')
ax[0].set_xlim(max(top10_countries_groupby['Total shows'])*1.1)
#ax[0].tick_params(left = False)
for spine in ax[0].spines.values():
    spine.set_visible(False)

# Percentage chart
ax[1].barh(width= 'Contribution', y = 'country', 
        data = top10_countries_groupby[top10_countries_groupby['type']=='Movie'].sort_values('Total shows', ascending=True), 
        left=top10_countries_groupby[top10_countries_groupby['type']=='TV Show'].sort_values('Total shows', ascending=True)['Contribution'], 
        color = '#221f1f',
        label = 'Movie')

ax[1].barh(width = 'Contribution', y = 'country', 
        data = top10_countries_groupby[top10_countries_groupby['type']=='TV Show'].sort_values('Total shows', ascending=True),
        color = '#b20710',
        label = 'TV shows')
#ax[1].set_title('Distribution of Movies & TV Shows across Top 10 Countries')
ax[1].set_xticks([])
new_lbl = [lbl.get_text().replace(' ', '\n') for lbl in ax[1].get_yticklabels()]

ax[1].set_yticks(ax[1].get_yticks())
ax[1].set_yticklabels(new_lbl)

for label in ax[1].get_yticklabels():
    label.set_ha('center')
    label.set_x(-0.085)  
    label.set_fontsize(14)
    label.set_weight('bold')
    label.set_color("#51191c")
    
ax[1].tick_params(left = False)

percentage_tv_show = top10_countries_groupby[top10_countries_groupby['type']=='TV Show'].sort_values('Total shows', ascending=True)['Contribution']
for i, v in enumerate(percentage_tv_show):
    ax[1].text(v/2+2, i, str(round(v))+"%", va='center', ha='right', fontsize=11, c = 'white')

percentage_movie = top10_countries_groupby[top10_countries_groupby['type']=='Movie'].sort_values('Total shows', ascending=True)['Contribution']
for i, v in enumerate( percentage_movie):
    ax[1].text(v/2 + percentage_tv_show.iloc[i], i, str(round(v))+"%", va='center', ha='right', fontsize=11, c = 'white')
    
for spine in ax[1].spines.values():
    spine.set_visible(False)

plt.figtext(x=0.16 , y= 0.33, s = 'Movie', fontdict = {'color':'#221f1f','fontsize':30,'weight':'bold'})
plt.figtext(x=0.16 , y= 0.25, s = 'TV Show', fontdict = {'color':'#b20710','fontsize':30,'weight':'bold'})

plt.figtext(x=0.15 , y= 0.88, s = 'Quantity and Proportion of Movies & TV Shows \nin Top 10 Production Countries', fontdict = {'color':"#51191c",'fontsize':25,'weight':'bold'})


plt.show()

```

## 3. Target Audiences

Overall, adult and teenager are the main user segments of Netflix.

There is a slight difference in target audiences among top 10 show making countries. Meanwhile, most countries focus on contents for adults, Asian countries such as India, South Korea, and Japan aim their target on teenagers.

```python

ta_type_groupby = data[['target_audience','type']][data['target_audience'] != 'Unrated'].value_counts().reset_index(name = 'No. shows')


fig, ax = plt.subplots(1, 1, figsize = (7, 3))

ta_type_pivot = ta_type_groupby.pivot(index = 'target_audience', columns= 'type', values = 'No. shows')
ta_type_pivot['Total'] = ta_type_pivot['Movie'] + ta_type_pivot['TV Show']
ta_type_pivot = ta_type_pivot.sort_values('Total', ascending=True)
ax.barh(width = ta_type_pivot['Movie'],
        y = ta_type_pivot.index.to_list(),
        color = '#221f1f',
        left = ta_type_pivot['TV Show']
)
ax.barh(width = ta_type_pivot['TV Show'],
        y = ta_type_pivot.index.to_list(),
        color = '#b20710'
)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(left = False, bottom = False)
ax.set_xticks([])
# ax.grid(axis = 'x', color = 'grey', linestyle = '--', linewidth = 0.5, alpha = 0.5)
# ax.set_xticks([])
for y, x in enumerate(ta_type_pivot['Total'].to_list()):
        ax.text(x + 50, y - 0.09, x, color = 'black', fontsize = 10 )

for label in ax.get_yticklabels():
    # label.set_ha('left')
    #label.set_x(-0.21)  
    #label.set_fontsize(14)
    label.set_weight('bold')
    label.set_color("#51191c")
fig.text(x = -0.08, y = 0.9, s='Movies & TV Shows by Target Audiences', fontdict = {'color':"#51191c",'fontsize':20,'weight':'bold'})
plt.figtext(x=0.66 , y= 0.33, s = 'Movie', fontdict = {'color':'#221f1f','fontsize':17,'weight':'bold'})
plt.figtext(x=0.66 , y= 0.23, s = 'TV Show', fontdict = {'color':'#b20710','fontsize':17,'weight':'bold'})
plt.show()
```

```python

xploded_ta_ctr_data = data[['target_audience','country']][data['target_audience'] != 'Unrated'].explode('country')
ta_ctr_groupby = xploded_ta_ctr_data.value_counts().reset_index(name = 'No. shows')
ta_ctr_pivot = ta_ctr_groupby.pivot(index='country', columns='target_audience',values='No. shows')
ta_ctr_pivot['Total'] = ta_ctr_pivot['Adults']+ta_ctr_pivot['General Audience'] + ta_ctr_pivot['Kids'] + ta_ctr_pivot['Teenagers']

ta_ctr_pivot = ta_ctr_pivot.sort_values('Total', ascending = False).head(10).sort_values('Total', ascending=True)

target_audience_filtered = ['Adults','Teenagers','Kids','General Audience']
for i in target_audience_filtered:
    ta_ctr_pivot[i] = ta_ctr_pivot[i]/ta_ctr_pivot['Total']*100


fig, ax = plt.subplots(figsize = (10,5))

colors = ['#221f1f','#b20710',"#979797","#ff7279"]

left = pd.Series([0]*len(ta_ctr_pivot.index.to_list()))


for i in range(len(target_audience_filtered)):

    ax.barh(y = ta_ctr_pivot.index.to_list(),
            width= ta_ctr_pivot[target_audience_filtered[i]],
            height= 0.8,
            left = left, label=target_audience_filtered[i],
            color = colors[i]
            )
    left = left + ta_ctr_pivot[target_audience_filtered[i]].values

ax.tick_params(left = False, bottom = False)
ax.set_xticks([25,50,75,100])
ax.set_xticklabels(['25%','50%','75%','100%'])

for spine in ax.spines.values():
    spine.set_visible(False)
for label in ax.get_yticklabels():
    # label.set_ha('left')
    #label.set_x(-0.21)  
    #label.set_fontsize(14)
    label.set_weight('bold')
    label.set_color("#51191c")

ax.set_title('Proportion of Target Audiences \nacross Top 10 Production Countries', 
             fontdict = {'color':"#51191c",'fontsize':25,'weight':'bold', 'ha':'left'},
             loc='left',
             x=-0.1, y=1.1)
plt.legend(ncols=4, 
           fontsize = 10,
           framealpha = 0, loc = 'upper left',  bbox_to_anchor=(0, 1.05)
           )
plt.show()

```

## 4. Show Genre

Among movies and TV shows, top 3 show categories were internationality, drama, and comedy. The same result happened with adults and teenagers.


It seems to be understandable that most of the shows on Netflix were published globally to engage as many viewers as possible, therefore, internationality is one of crucial qualities film producers would concentrate on. In addition, drama and comedy are critical materials for building a thrilling script to keep subscriber engaged and excited. 


```python

genre_data = data[['listed_in','type']][data['listed_in'].notna()].explode('listed_in').value_counts().reset_index(name='No. shows')
genre_data_excl_international = genre_data[~genre_data['listed_in'].str.lower().str.contains('international')]

top7_genre = genre_data.sort_values('No. shows', ascending=False)
top7_genre_xcl_int = genre_data_excl_international.sort_values('No. shows', ascending=False)

fig, axs = plt.subplots(1, 2, figsize= (15,6))

type_n_colors = {'Movie':'#221f1f','TV Show':'#b20710'}



axs[0].barh(width= top7_genre[top7_genre['type'] == 'Movie'].head(7)['No. shows'],
            y= top7_genre[top7_genre['type'] == 'Movie'].head(7)['listed_in'].values,
            color = [type_n_colors['Movie']],
            )
axs[0].set_xticks([1000,2000,3000])
axs[1].barh(width= top7_genre[top7_genre['type'] == 'TV Show'].head(7)['No. shows'],
            y= top7_genre[top7_genre['type'] == 'TV Show'].head(7)['listed_in'].values,
            color = [type_n_colors['TV Show']],
            )
axs[1].set_xticks([500,1000,1500])


for ax in axs.flat:
    ax.invert_yaxis()
    ax.tick_params(bottom=False, left = False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    for label in ax.get_yticklabels():
        label.set_weight('bold')
        label.set_color("#51191c")


axs[0].set_title('Top 7 Genres of Movie & TV Show',
                fontdict = {'color':"#51191c",'fontsize':30,'weight':'bold', 'ha':'left'},
                loc='left',
                #x=-0.1, y=1.1
                )
plt.figtext(x=0.33 , y= 0.33, s = 'Movie', fontdict = {'color':'#221f1f','fontsize':30,'weight':'bold'})
plt.figtext(x=0.83 , y= 0.33, s = 'TV Show', fontdict = {'color':'#b20710','fontsize':30,'weight':'bold'})
plt.tight_layout()
plt.show()

```

```python

genre_ta_data = data[['listed_in','target_audience']][(data['target_audience'] != 'Unrated') & (data['listed_in'].notna())]
xploded_genre_ta = genre_ta_data.explode('listed_in')
xploded_genre_ta['listed_in'] = xploded_genre_ta['listed_in']\
                                .str.replace('Movies', '')\
                                .str.replace('TV Shows', '')\
                                .str.replace('TV', '')\
                                .str.strip()
grb_genre_ta = xploded_genre_ta.value_counts().reset_index(name = 'No. shows')
pivot_genre_ta = grb_genre_ta.pivot(index='listed_in', columns = 'target_audience', values = 'No. shows')

tar_audience = ['Adults','Teenagers']
colors1 = ['#221f1f','#b20710']

fig, axs = plt.subplots(1, 2, figsize = (15,6))

for i in range(2):
    get_data = pivot_genre_ta[tar_audience[i]].sort_values(ascending=False).head(7)
    axs[i].barh(width = get_data,
               y = get_data.index.to_list(),
               color = colors1[i])
    for spine in axs[i].spines.values():
        spine.set_visible(False)

    axs[i].tick_params(bottom = False, left =False)
    axs[i].invert_yaxis()
    axs[i].text(1200, 4, 
                tar_audience[i], 
                color = colors1[i], 
                fontsize = 30,
                weight = 'bold')
    for label in axs[i].get_yticklabels():
        label.set_weight('bold')
        label.set_color("#51191c")
    axs[i].set_xticks([500,1000,1500,2000])
axs[0].set_title('Top 7 Genres of Movie & TV Show',
                fontdict = {'color':"#51191c",'fontsize':30,'weight':'bold', 'ha':'left'},
                loc='left',
                #x=-0.1, y=1.1
                )
plt.tight_layout()
plt.show()
```

# CONCLUSION

Regarding the insights explored, non-Asian countries in the top 10 exhibit a preference for film production over TV, with a primary focus on adult audiences.

By contrast, Asian leaders such as South Korea and Japan concentrate on TV series aimed at teenagers.

India, however, shows a notable exception; it maintains a film-heavy output (90%) while simultaneously targeting a younger segment, indicating a mix of Asian and non-Asian characteristics.

Despite these regional differences above, there is a global trend toward utilizing international themes, drama, and comedy in scriptwriting.
